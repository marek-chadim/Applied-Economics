```python
    import pandas as pd
    import numpy as np

    data_url = "http://lib.stat.cmu.edu/datasets/boston"
    bos = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
    data = np.hstack([bos.values[::2, :], bos.values[1::2, :2]])


boston = pd.DataFrame(data)
boston['MEDV'] = bos.values[1::2, 2]
boston.head()
```

```python
%%stata -d boston
describe
```

```python
%%stata
regress MEDV v1 v2 v3
```
